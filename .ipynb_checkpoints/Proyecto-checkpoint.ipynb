### Funciones para eliminar palabras inutiles, simbolos y para insertar en dataframe cada palabra con su conteo

```python
prep = ['is','and','the','a',"it's", 'his','her','has','with','on','my','not','host','people','an','to','or','no','in','of','you','for','room','are','very','super','fast','can','are','amsterdam','city','&','minutes','from','this']
rplc =  ['(',')',',','.',':','-','"',"'"]
def limpiarPalabras(description):
    arr = description.split(" ")
    newarr = []
    for palabra in arr:
        if palabra not in prep:
            for letra in palabra:
                if letra in rplc:
                    palabra = palabra.replace(letra,"")
            newarr.append(palabra)
    return newarr
    


def nuevoDF(ncolumna):
    w = []
    c = []
    for i in range(0,len(lisd)):
        description = lisd.iloc[i,ncolumna]
        if(type(description)==str):
            description = description.lower()
            newarr = limpiarPalabras(description)
            for palabra in newarr:
                if palabra not in w:
                    w.append(palabra)
                    c.append(1)
                else:
                    index = w.index(palabra)
                    c[index] +=1
    return regresarDF(w,c)

def regresarDF(w,c):
    col = ['w','c']
    palabras = pd.DataFrame(columns = col)
    palabras['w']=w
    palabras['c']=c
    return palabras
```

### Codigo para limpiar amenities y contar cuantas veces se ofrecen ciertos servicios como wifi, calentadores, etc

```python
def limpiarAmenities():
    newarr = []
    for i in range(0,len(lisd)):
        string = lisd.iloc[i,14].lower()
        arreglo = string.split(",")
        remplazar = '}{"'
        for palabra in arreglo:
            for char in remplazar:
                palabra = palabra.replace(char,"")
            newarr.append(palabra)
    return newarr
    

    
def nuevoDFAmenities():
    w = []
    c = []
    newarr = limpiarAmenities()
    for palabra in newarr:
        if palabra not in w:
            w.append(palabra)
            c.append(1)
        else:
            index = w.index(palabra)
            c[index] +=1
    return regresarDFAmenities(w,c)

def regresarDFAmenities(w,c):
    col = ['w','c']
    palabras = pd.DataFrame(columns = col)
    palabras['w']=w
    palabras['c']=c
    return palabras
```

# Now let's try a Neural Network

```python
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
def KerasNN():
    model = Sequential()
    model.add(Dense(5, input_dim=28, activation='relu'))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```

```python
# fix random seed for reproducibility
seed = 43
np.random.seed(seed)

estimator = KerasRegressor(build_fn=KerasNN, epochs=100, batch_size=5000, verbose=0)
```

```python
def deepNN():
    model = Sequential()
    model.add(Dense(5, input_dim=5, kernel_initializer='normal', activation='relu'))
    model.add(Dense(20, kernel_initializer='normal', activation='relu'))
    model.add(Dense(10, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```
